```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
import math
%matplotlib inline
```

```python
def light ():
    slab = [2,-2]
    photons = []
    c_refracted = 0
    c_transmitted = 0
    c_absorbed = 0
    for i in range(100):

        photon = []
        position = np.array( [0.0,15] ) #x,y
        direction = np.array( [0.0,-1.0] ) #x,y
        refracted = False
        transmitted = False
        absorbed = False

        while not refracted and not transmitted and not absorbed:
            photon.append(position)
            #print position

            if position[1] <= slab[0] and position[1] >= slab[1]: #inside
                #print 'inside'

                absorbed = np.random.randint(100)%99 == 0
                if absorbed:
                    c_absorbed = c_absorbed + 1
                    absorbed = True
                else:
                    refract = np.random.rand()*2.0*math.pi # 2*Pi*Random radians
                    magnitude = np.sqrt(direction.dot(direction))
                    direction = np.array( [magnitude*math.cos(refract), magnitude*math.sin(refract)] )
                    distance = np.random.exponential()
                    position = np.array( position + direction*distance )

            else: #outside
                if direction[1] > 0: #refracted
                    c_refracted = c_refracted + 1
                    refracted = True

                elif direction[1] <= 0 and position[1] < slab[1]: #transmitted
                    c_transmitted = c_transmitted + 1
                    transmitted = True

                else:
                    #not entered in the slab yet
                    position = np.array( position + direction )
            pass
        photons.append(photon)
    return c_refracted, c_transmitted, c_absorbed, photons
```

```python
d_refracted = []
d_transmitted = []
d_absorbed = []

for i in range(10):
    results = light()
    d_refracted.append(results[0])
    d_transmitted.append(results[1])
    d_absorbed.append(results[2])
```

# A single sample visualized

```python
plt.plot()
slab = [2,-2]
plt.axhline(slab[0])
plt.axhline(slab[1])
plt.title("Light transmitting through a slab")
photons = results[3]
for photon in photons:

    photon_x = []
    photon_y = []
    for pos in photon:
        photon_x.append(pos[0])
        photon_y.append(pos[1])

    plt.plot(photon_x,photon_y,lw=0.3)


plt.xticks(np.arange(-10.0,10.1,1.0))
plt.show()
```

# Histograms

```python
f, axarr = plt.subplots(1,3,sharey=True, figsize=(15,5))


axarr[0].set_title('Refracted')
mu, std = norm.fit(d_refracted)
x = np.linspace( mu-4*std , mu+4*std ,100)
y = norm.pdf(x,mu,std)

axarr[0].plot(x,y,"r-",lw=1.5)

n, bins, _ = axarr[0].hist(d_refracted, normed=True, alpha=0.8)
mid = 0.5*(bins[1:] + bins[:-1])
axarr[0].errorbar(mid, n, yerr=np.sqrt(n)*n, fmt='none')



axarr[1].set_title('Transmitted')
mu, std = norm.fit(d_transmitted)
x = np.linspace( mu-4*std , mu+4*std ,100)
y = norm.pdf(x,mu,std)

axarr[1].plot(x,y,"r-",lw=1.5)

n, bins, _ = axarr[1].hist(d_transmitted, normed=True, alpha=0.8)
mid = 0.5*(bins[1:] + bins[:-1])
axarr[1].errorbar(mid, n, yerr=np.sqrt(n)*n, fmt='none')


axarr[2].set_title('Absorbed')
mu, std = norm.fit(d_absorbed)
x = np.linspace( mu-4*std , mu+4*std ,100)
y = norm.pdf(x,mu,std)

axarr[2].plot(x,y,"r-",lw=1.5)

n, bins, _ = axarr[2].hist(d_absorbed, normed=True, alpha=0.8)
mid = 0.5*(bins[1:] + bins[:-1])
axarr[2].errorbar(mid, n, yerr=np.sqrt(n)*n, fmt='none')




axarr[0].grid(True)
axarr[1].grid(True)
axarr[2].grid(True)

plt.show()
```
