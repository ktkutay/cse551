```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
def get_sa(r):
    np.random.seed(1)
    b = np.random.uniform(-r,r,100000) #100k
    sa = 2 * np.arccos(b/r) * 180 / np.pi # scattering angles
    return sa[~np.isnan(sa)]
```

```python
fig, ax = plt.subplots(2,2,figsize=(9,6))
r = [1,2,4,8]
c = 0
for i in range(2):
    for j in range(2):
        ax[i][j].hist(get_sa(r[c]),bins=20,normed=True)
        ax[i][j].set_title( "".join(["$r=",str(r[c]),"$"]))
        ax[i][j].set_ylabel('$P$')
        ax[i][j].set_xlabel('scattering angle')
        c+=1
plt.tight_layout()
plt.show()
```
